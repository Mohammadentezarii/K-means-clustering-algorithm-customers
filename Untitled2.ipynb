```python
data["A"] = data[["Annual Income (k$)"]]
data["B"] = data[["Spending Score (1-100)"]]
```

```python
X = data[["A","B"]]
X.head()
```

```python
# @title A vs B
X.plot(kind='scatter', x='A', y='B', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
Kmean = KMeans(n_clusters = 5)
Kmean.fit(X)
```

```python
centers = Kmean.cluster_centers_
print(centers)
```

```python
clusters = Kmean.fit_predict(X)
data["label"] = clusters
data.head(100)
```

```python
col = ['green', 'blue', 'black', 'yellow', 'orange']

for i in range(5):
  a = col[i]
  plt.scatter(data.A[data.label == i], data.B[data.label == i], c = a, label = 'cluser 1')
plt.scatter(centers[:, 0], centers[:, 1], marker = '*', s = 300, c = 'r', label = 'centroid')
```

```python
X1 = X.loc[:,["A","B"]].values
```

```python
wcss = []
for K in range(1, 11):
    Kmeans = KMeans(n_clusters=K, init="k-means++")
    Kmeans.fit(X1)
    wcss.append(Kmeans.inertia_)

plt.figure(figsize=(12, 6))
plt.grid()
plt.plot(range(1, 11), wcss, linewidth = 2, color='red', marker='8')
plt.xlabel('K Value')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show()
```
